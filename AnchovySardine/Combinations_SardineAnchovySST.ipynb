Notebook to generate combination of array to run analysis on.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
```

```python
SASST = pd.read_csv('sardine_anchovy_sst.csv')
```

```python
SASST
```

```python
SASST = SASST.drop(labels = ['year'], axis = 1)
SASST
```

```python
names = SASST.columns.values.tolist()
names
```

```python
combs = list(combinations(names,2))
combs
```

```python
np.savetxt('combs_sardineanchovy.csv', combs, delimiter = ',', fmt ='% s')
```
