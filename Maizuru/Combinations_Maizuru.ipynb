Notebook to generate combination of array to run analysis on.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
```

```python
maizuru = pd.read_csv('Maizuru_dominant_sp.csv')
maizuru
```

```python
maizuru.drop(['date_tag', 'surf.t', 'bot.t', 'Y', 'M', 'D'], inplace = True, axis = 1)
maizuru
```

```python
names = maizuru.columns.values.tolist()
names
```

```python
combs = list(combinations(names,2))
combs
```

```python
np.savetxt('combs_maizuru.csv', combs, delimiter = ',', fmt ='% s')
```
