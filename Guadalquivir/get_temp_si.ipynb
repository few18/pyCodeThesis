```python
import numpy as np
import pandas as pd
```

```python
env = pd.read_csv('GEnv_NOgaps.csv')
env
```

```python
env_small = env[['Date5.1', 'Site', 'Tide', 'Temp', 'Si']]
```

```python
env_small['Date5.1'] = pd.to_datetime(env_small["Date5.1"])
```

```python
env_small = env_small.loc[:, ~env_small.columns.str.contains("^Unnamed")]
```

```python
env_grouped = env_small.groupby(["Date5.1"], as_index=False).sum()
```

```python
env_grouped = env_grouped.drop(["Site", "Tide"], axis=1)
```

```python
env_grouped
```

```python
env_grouped = env_grouped.fillna(method="bfill") 
```
